```python
folder_path = "drive/MyDrive/train_img"
folder_filenames = os.listdir(folder_path)
folder_filenames.sort()
print(folder_filenames)
img = None
soup = None
right = 0
count = 0
def get_lines(img, kernel, erode_iterations = 1, dilate_iterations = 1):
  erode_img = cv2.erode(img, kernel, erode_iterations)
  lines = cv2.dilate(erode_img, kernel, dilate_iterations)
  return lines
def cordinates_line(lines):
  first_cordinates = cv2.HoughLinesP(lines, 1, np.pi/180, 30, maxLineGap=250)
def contour_boxes(img):
  contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  return  np.array([[x, y, x + w, y + h] for contour in contours for (x, y, w, h) in [cv2.boundingRect(contour)] if w > 5 and h > 5]), contours
def concatenate_lines(lines, VoH, eps):
  copy_lines = lines.copy()
  new_lines = []
  copy_lines = copy_lines[np.argsort(copy_lines[:, VoH])]
  line_equations = []
  while copy_lines.shape[0] > 0:
    if copy_lines.shape[0] == 1:
      new_lines.append(copy_lines[0])
      break
    value1 = copy_lines[0, VoH]
    check_value1 = np.abs(copy_lines[:, VoH] - value1) <= eps
    index = (VoH + 1) % 2
    value2 = min(copy_lines[0, index], copy_lines[0, index + 2])
    value3 = max(copy_lines[0, index + 2], copy_lines[0, index + 2])
    concatenate_value = np.hstack((copy_lines[:, index].reshape(-1, 1), copy_lines[:, index + 2].reshape(-1, 1)))
    min_value = np.min(concatenate_value, axis = 1)
    max_value = np.max(concatenate_value, axis = 1)
    check_value2 = min_value >= value2
    check_value3 = min_value <= value3
    check_value4 = max_value >= value2
    check_value5 = min_value <= value2
    check = check_value1 & ((check_value2 & check_value3) | (check_value4 & check_value5))
    same_lines = copy_lines[check]
    min_value = min_value[check]
    max_value = max_value[check]
    copy_lines = copy_lines[check == False]
    if VoH == 0:
      new_lines.append(np.array([np.mean(same_lines[:, VoH]), np.min(min_value), np.mean(same_lines[:, VoH]), np.max(max_value)]))
    else:
      new_lines.append(np.array([np.min(min_value), np.mean(same_lines[:, VoH]), np.max(max_value), np.mean(same_lines[:, VoH])]))
  return np.array(new_lines), line_equations
#np.concatenate((v_lines, h_lines))
def area(boxes):
  return (boxes[:, 2] - boxes[:, 0] + 1.) * (boxes[:, 3] - boxes[:, 1] + 1.)
def iou(one_box, boxes):
  one_boxes = np.full((boxes.shape[0],) + one_box.shape, one_box)
  overlapped_boxes = one_boxes.copy()
  for i in range(4):
    concatenate_values = np.hstack((one_boxes[:, i].reshape(-1, 1), boxes[:, i].reshape(-1, 1)))
    if i < 2:
      overlapped_boxes[:, i] = np.max(concatenate_values, axis = 1)
    else:
      overlapped_boxes[:, i] = np.min(concatenate_values, axis = 1)
  overlapped_s = area(overlapped_boxes)
  overlapped_s[((overlapped_boxes[:, 2] - overlapped_boxes[:, 0] < 0) | (overlapped_boxes[:, 3] - overlapped_boxes[:, 1] < 0))] = 0.
  two_s = area(one_boxes) + area(boxes)
  return (overlapped_s / (two_s - overlapped_s)), overlapped_s, area(one_boxes), area(boxes)
def draw_boxes(img, boxes, color, thickness):
  img_copy = img.copy()
  for box in boxes:
    x1 = int(box[0])
    y1 = int(box[1])
    x2 = int(box[2])
    y2 = int(box[3])
    cv2.rectangle(img_copy, (x1,y1), (x2,y2), color, thickness)
  return img_copy
def draw_boxes2(img, boxes, color, thickness):
  img_copy = img.copy()
  for box in boxes:
    x1 = int(box[0])
    y1 = int(box[1])
    x2 = int(box[2])
    y2 = int(box[3])
    img_copy[y1 : y2, x1 : x1 + int(thickness)] = color
    img_copy[y1 : y2, x2 : x2 + int(thickness)] = color
    img_copy[y1 : y1 + int(thickness), x1 : x2] = color
    img_copy[y2 : y2 + int(thickness), x1 : x2] = color
  return img_copy
def concatenate_boxes(text_boxes, eps1, eps2, VoH):
  sort_boxes = text_boxes[np.argsort(text_boxes[:, VoH])]
  new_boxes = []
  while sort_boxes.shape[0] > 0:
    if sort_boxes.shape[0] == 1:
      new_boxes.append(sort_boxes[0])
      break
    one_box = sort_boxes[0]
    iou_boxes, overlapped_s, one_s, rest_s = iou(one_box, sort_boxes)
    boxes_overlapped = sort_boxes[iou_boxes >= eps1]
    sort_boxes = sort_boxes[iou_boxes < eps1]
    def update_box(one_box, boxes):
      box_copy = []
      for i in range(4):
        if i < 2:
          box_copy.append(np.min(boxes[:, i]))
        else:
          box_copy.append(np.max(boxes[:, i]))
      return np.array(box_copy)
    one_box = update_box(one_box, boxes_overlapped)
    sort_boxes2 = sort_boxes.copy()
    sort_boxes2[:, VoH] -= eps2 + 2
    sort_boxes2[:, VoH + 2] -= eps2 + 2
    sort_boxes[sort_boxes[:, VoH] < 0.][:, VoH] = 0.
    sort_boxes[sort_boxes[:, VoH + 2] < 0.][:, VoH + 2] = 0.
    iou_boxes, overlapped_s, one_s, rest_s = iou(one_box, sort_boxes2)
    min_s = np.min(np.hstack((one_s.reshape(-1, 1), rest_s.reshape(-1, 1))), axis = 1)
    check2 = (iou_boxes > 0) & (np.abs(sort_boxes[:, VoH] - one_box[VoH + 2]) <= eps2) & ((overlapped_s - min_s) != 0.)
    boxes_near = sort_boxes[check2]
    sort_boxes = sort_boxes[check2 == False]
    boxes_near = np.concatenate((one_box.reshape(-1, 4), boxes_near))
    one_box = update_box(one_box, boxes_near)
    new_boxes.append(one_box)
  return np.array(new_boxes)
def frequency_color(img):
  frequency = np.array([img[img == i].shape[0] for i in range(256)])
  return np.argmax(frequency)
def draw_line(lines, img, color, thickness):
    img_copy = img.copy()
    for line in all_lines:
      x1, y1, x2, y2 = line.astype(int)
      cv2.line(img_copy, (x1, y1), (x2, y2), float(color), thickness)
    return img_copy
id = "/72"
for index, filename in enumerate(folder_filenames):
  #if index > 2: break
  if index % 2 == 0:
    img = Image.open(os.path.join(folder_path, filename)).convert("RGB")
    #folder_filename = folder_path + id + ".png"
    #img = Image.open(folder_filename).convert("RGB")
  else:
    folder_filename = os.path.join(folder_path, filename)
    #folder_filename = folder_path + id + ".xml"
    with open(folder_filename, 'r') as f:
      CONTENTS = f.read()
    soup = BeautifulSoup(CONTENTS, 'xml')
    w = int(soup.width.string)
    h = int(soup.height.string)
    bboxes = []
    for o in soup.find_all("object"):
      xMin = int(o.find("xmin").string)
      yMin = int(o.find("ymin").string)
      xMax = int(o.find("xmax").string)
      yMax = int(o.find("ymax").string)
      bboxes.append([xMin, yMin, xMax, yMax])
    bboxes = np.array(bboxes, dtype=np.float32)
    img = np.array(img)
    img_copy = img.copy()
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_canny  = cv2.Canny(img , 100, 200)
    img_canny = cv2.threshold(img_canny, 127, 255, cv2.THRESH_BINARY)[1]
    #plt.imshow(img_canny)
    v_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (40, 1))
    text_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (12, 1))
    h_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 40))
    get_vlines = get_lines(img_canny, v_kernel, 1, 1)
    get_hlines = get_lines(img_canny, h_kernel, 1, 1)
    v_lines = np.array([])
    try:
      v_lines = cv2.HoughLinesP(get_vlines, 1, np.pi/180, 30, maxLineGap=250).reshape(-1, 4)
    except:
      v_lines = v_lines.reshape(-1, 4)
    v_lines, vline_equations = concatenate_lines(v_lines, 1, 5)
    try:
      h_lines = cv2.HoughLinesP(get_hlines, 1, np.pi/180, 30, maxLineGap=250).reshape(-1, 4)
    except:
      h_lines = v_lines.reshape(-1, 4)
    h_lines, hline_equations = concatenate_lines(h_lines, 0, 5)
    all_lines = np.concatenate((v_lines, h_lines))
    all_equations = np.concatenate((vline_equations, hline_equations))
    all_lines = np.concatenate((all_lines, np.array([[0, 0, img_canny.shape[1], img_canny.shape[0]]])))
    full_255 = img_canny.copy()
    full_255[:, :] = 255.
    full_255 = draw_line(all_lines, full_255, 0, 5)
    cell_boxes, contours = contour_boxes(full_255)
    color = frequency_color(img)
    img = draw_line(all_lines, img, color, 5)
    ret, img = cv2.threshold(img, 127,255, cv2.THRESH_BINARY)
    #plt.imshow(img)
    #plt.imshow(full_255)
    get_texts = get_lines(img, text_kernel, 1, 1)
    get_texts[:-1, :5] = 255.
    get_texts[:-1, -10:-1] = 255.
    get_texts[:5, :-1] = 255.
    text_boxes, contours = contour_boxes(get_texts)
    all_cells = text_boxes.copy()
    text_boxes = concatenate_boxes(text_boxes, 0.3, 10, 0)
    text_boxes = concatenate_boxes(text_boxes, 0.3, 10, 1)
    all_cells = np.concatenate((all_cells, text_boxes))
    for box in bboxes:
      count += 1
      iou_box = iou(box,  all_cells)[0]
      if iou_box[iou_box > 0.5].shape[0] > 0:
        right += 1
    #img_copy = draw_boxes(img_copy, all_cells, (255, 0, 0), 2)
    img_copy = draw_boxes(img_copy, all_cells, (255, 0, 0), 2)
    if filename == "48811.xml": plt.imshow(img_copy)
print(right / count)
```
